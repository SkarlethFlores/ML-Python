```python
# Ejercicio
# - train-test split
# - chi2, os quedais con 3..10 variables (bucle)
# - minmaxscaler
# - logistic regression
# ---> en pipeline

from sklearn.pipeline import Pipeline, make_pipeline

X = df[num_feats]
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

accuracies = []
for i in range(3, 31):

#     pipeline = make_pipeline(selector, scaler, model)
    pipeline = Pipeline([
        ('selector-edu', SelectKBest(score_func=chi2, k=i)),
        ('minmax', MinMaxScaler()),
        ('LR', LogisticRegression())
    ])
    
    pipeline.fit(X_train, y_train)

    acc = pipeline.score(X_test, y_test)
#     print(f'acc={acc:.3f}, i={i}')
    
    accuracies.append(acc)
    
plt.figure(figsize=(10,2))
plt.plot(accuracies)
plt.show()
```

```python
""" model selection """

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

models = []

models.append(('LR', LogisticRegression(random_state=99)))
models.append(('DTC', DecisionTreeClassifier(random_state=99)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('NB', GaussianNB()))
models.append(('RFC', RandomForestClassifier(random_state=99)))
models.append(('SVC', SVC(random_state=99)))

results = []
names = []
for name, model in models:
    scaler = StandardScaler()
    pipeline = make_pipeline(scaler, model)
    kf = KFold(n_splits=10, shuffle=True, random_state=99)
    acc = cross_val_score(pipeline, X, y, cv=kf)
    
    names.append(name)
    results.append(acc)

plt.figure(figsize=(8,4))
plt.boxplot(results)
plt.xticks(range(1, len(names)+1), names)
plt.show()
```

# Model tuning

```python
from sklearn.model_selection import GridSearchCV

model = DecisionTreeClassifier()

parametros = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [3,6,12,100]
}

kf = KFold(n_splits=5)

grid_model = GridSearchCV(
    estimator=model,
    param_grid=parametros,
    cv=kf
)
grid_model.fit(X, y)

grid_model.best_estimator_
grid_model.best_score_
grid_model.predict(X)
grid_model.score(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()

parametros = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [3,6]
}

kf = KFold(n_splits=5)

grid_model = GridSearchCV(
    estimator=model,
    param_grid=parametros,
    cv=kf
)

# training

grid_model.fit(X_train, y_train)

# evaluating

y_hat = grid_model.predict(X_train)
acc_train = accuracy_score(y_true=y_train, y_pred=y_hat)

y_hat = grid_model.predict(X_test)
acc_test = accuracy_score(y_true=y_test, y_pred=y_hat)

print(acc_test)
print(acc_train)
```
